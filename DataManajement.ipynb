```python
class Mahasiswa:
    def __init__(self, nama, nim, kelas, asal, gender):
        self.nama = nama
        self.nim = nim
        self.kelas = kelas
        self.asal = asal
        self.gender = gender
    def __str__(self):
        return f"{self.nama},{self.nim},{self.kelas},{self.asal},{self.gender}"

class DatabaseMahasiswa:
    def __init__(self, file_name):
        self.mahasiswa_list = []
        self.file_name = file_name
        self.load_data()
    def load_data(self):
        try:
            with open(self.file_name, "r") as file:
                for line in file:
                    if line.strip():
                        data = line.strip().split(",")
                        if len(data) >= 3:
                            nama = data[0]
                            nim = data[1]
                            kelas = data[2]
                            asal = data[3]
                            gender = data[4]
                            self.mahasiswa_list.append(Mahasiswa(nama, nim, kelas, asal, gender))
        except FileNotFoundError:
            pass
    def save_data(self):
        with open(self.file_name, "w") as file:
            for mahasiswa in self.mahasiswa_list:
                file.write(f"{mahasiswa}\n")

    def tampilkan_data(self):
        if not self.mahasiswa_list:
            print("Database kosong!")
            return
        print("\n" + "="*71)
        print(f"{'Nama':<20} {'NIM':<15} {'Kelas':<10} {'asal':<15} {'gender':<10}")
        print("-"*71)
        for mahasiswa in self.mahasiswa_list:
            print(f"{mahasiswa.nama:<20} {mahasiswa.nim:<15} {mahasiswa.kelas:<10} {mahasiswa.asal:<15} {mahasiswa.gender:<10}")
        print("="*71)
        print(f"Total: {len(self.mahasiswa_list)} mahasiswa\n")
    
    def tambah_data(self, nama, nim, kelas, asal, gender):
        for mahasiswa in self.mahasiswa_list:
            if mahasiswa.nim == nim:
                print(f"Mahasiswa dengan NIM {nim} sudah ada dalam database!")
                return False
        self.mahasiswa_list.append(Mahasiswa(nama, nim, kelas, asal, gender))
        self.save_data()
        print(f"Data mahasiswa {nama} berhasil ditambahkan!")
        return True
    
    def hapus_data(self, nim):
        for i, mahasiswa in enumerate(self.mahasiswa_list):
            if mahasiswa.nim == nim:
                deleted = self.mahasiswa_list.pop(i)
                self.save_data()
                print(f"Data mahasiswa {deleted.nama} dengan NIM {nim} berhasil dihapus!")
                return True
        print(f"Mahasiswa dengan NIM {nim} tidak ditemukan!")
        return False

    def cari_data(self, nim):
        for mahasiswa in self.mahasiswa_list:
            if mahasiswa.nim == nim:
                print("\nData ditemukan:")
                print(f"Nama  : {mahasiswa.nama}")
                print(f"NIM   : {mahasiswa.nim}")
                print(f"Kelas : {mahasiswa.kelas}")
                return True
        print(f"Mahasiswa dengan NIM {nim} tidak ditemukan!")
        return False

def main():
    db = DatabaseMahasiswa("Data.txt")
    while True:
        print("\nSistem Manajemen Data Mahasiswa")
        print("1. Tampilkan Data")
        print("2. Tambah Data")
        print("3. Hapus Data")
        print("4. Cari Data")
        print("0. Keluar")
        pilihan = input("Pilih menu (0-4): ")
        
        if pilihan == "1":
            db.tampilkan_data()
        elif pilihan == "2":
            nama = input("Masukkan nama: ")
            nim = input("Masukkan NIM: ")
            kelas = input("Masukkan kelas: ")
            asal = input("Masukkan asal: ")
            gender = input("Masukkan gender: ")
            db.tambah_data(nama, nim, kelas, asal, gender)
        elif pilihan == "3":
            nim = input("Masukkan NIM mahasiswa yang akan dihapus: ")
            db.hapus_data(nim)
        elif pilihan == "4":
            nim = input("Masukkan NIM mahasiswa yang akan dicari: ")
            db.cari_data(nim)
        elif pilihan == "0":
            break
        else:
            print("Pilihan tidak valid. Silakan coba lagi.")

if __name__ == "__main__":
    main()
```
